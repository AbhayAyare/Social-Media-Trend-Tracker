```python
# Data Collection Notebook
# This notebook demonstrates how to collect data from Twitter and Reddit


import sys
sys.path.append('../src')

from data_collection import DataCollector
import pandas as pd
from datetime import datetime

# Initialize data collector
collector = DataCollector()

# Define search queries
twitter_queries = [
    "AI OR ChatGPT OR OpenAI",
    "machine learning OR deep learning", 
    "data science OR analytics",
    "tech OR technology",
    "programming OR coding"
]

# Collect Twitter data
print("üïí Starting Twitter data collection...")
twitter_df = collector.scrape_twitter_data(
    queries=twitter_queries,
    days_back=7,
    max_tweets=1000
)

# Collect Reddit data
print("üïí Starting Reddit data collection...")
reddit_df = collector.scrape_reddit_data(
    subreddits=["technology", "MachineLearning", "artificial", "datascience"],
    days_back=7, 
    max_posts=500
)

print("‚úÖ Data collection complete!")
print(f"Twitter data: {len(twitter_df)} posts")
print(f"Reddit data: {len(reddit_df)} posts")
```
